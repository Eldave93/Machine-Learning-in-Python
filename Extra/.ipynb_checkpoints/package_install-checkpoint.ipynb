# Anaconda
I ended up using Pip but you can use the code below if you wanted to install everything with conda. I think is a bit less likely to cause package problems but takes longer.

```python
if CONDA:
    # Install a conda package in the current Jupyter kernel
    import sys
    print('scikit-learn')
    !conda install --yes --prefix {sys.prefix} scikit-learn
    print('pandas')
    !conda install --yes --prefix {sys.prefix} pandas
    print('gitpython')
    !conda install --yes --prefix {sys.prefix} gitpython
    print('graphviz')
    !conda install --yes --prefix {sys.prefix} graphviz
    print('imbalanced-learn')
    !conda install --yes --prefix {sys.prefix} imbalanced-learn
    print('dtreeviz')
    #!conda install --yes --prefix {sys.prefix} dtreeviz
    print('SKIP: Causes Issues')
    print('scipy')
    !conda install --yes --prefix {sys.prefix} scipy==1.6
    print('seaborn')
    !conda install --yes --prefix {sys.prefix} seaborn
    print('joblib')
    !conda install --yes --prefix {sys.prefix} joblib
    print('mlxtend')
    !conda install --yes --prefix {sys.prefix} mlxtend
    print('xlrd')
    !conda install --yes --prefix {sys.prefix} xlrd
    print('openpyxl')
    !conda install --yes --prefix {sys.prefix} openpyxl
    print('rise')
    !conda install --yes --prefix {sys.prefix} rise
```
